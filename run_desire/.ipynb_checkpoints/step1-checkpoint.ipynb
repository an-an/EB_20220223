# Read profile txt

```python
f = open(infile, 'r')
for iline in range(0,3):
    nwave  = int(f.readline())
    wave   = np.zeros((nwave), dtype = float)
    stki   = np.zeros((nwave), dtype = float)
    stkq   = np.zeros((nwave), dtype = float)
    stku   = np.zeros((nwave), dtype = float)
    stkv   = np.zeros((nwave), dtype = float)
    for iwave in range(0,nwave):
        txt = f.readline()
        txt1= (txt.split('\n'))[0]
        txt2= txt1.split(' ')
        txt3= [''] * 5 
        ipol = 0
        for i in range(0,len(txt2)):
            if txt2[i] != '':
                txt3[ipol] = float(txt2[i])
                ipol=ipol+1
        wave[iwave] = float(txt3[0])
        stki[iwave] = float(txt3[1])
        stkq[iwave] = float(txt3[2])
        stku[iwave] = float(txt3[3])
        stkv[iwave] = float(txt3[4])

    if iline == 0:
        wave_397 = wave
        stki_397 = stki
        stkq_397 = stkq
        stku_397 = stku
        stkv_397 = stkv
    if iline == 1:
        wave_630 = wave
        stki_630 = stki
        stkq_630 = stkq
        stku_630 = stku
        stkv_630 = stkv
    if iline == 2:
        wave_854 = wave
        stki_854 = stki
        stkq_854 = stkq
        stku_854 = stku
        stkv_854 = stkv

f.close
```

```python
print('Observed profile')
fig, ax = plt.subplots(nrows=12, ncols=1, figsize = (8,20), dpi=80)
#fig.suptitle('Observed profile', fontsize=16)
ax = ax.flatten()
for iline in range(0,3):
    if iline == 0:
        wavename = 'Ca II H & H epsilon'
        wave = wave_397
        stki = stki_397
        stkq = stkq_397
        stku = stku_397
        stkv = stkv_397
    if iline == 1:
        wavename = 'Fe I 630 nm'
        wave = wave_630
        stki = stki_630
        stkq = stkq_630
        stku = stku_630
        stkv = stkv_630
    if iline == 2:
        wavename = 'Ca II 854 nm'
        wave = wave_854
        stki = stki_854
        stkq = stkq_854
        stku = stku_854
        stkv = stkv_854

    for ipol in range(0,4):
        if ipol == 0:
            title = wavename+', Stokes I'
            stk = stki
        if ipol == 1:
            title = wavename+', Stokes Q'
            stk = stkq
        if ipol == 2:
            title = wavename+', Stokes U'
            stk = stku
        if ipol == 3:
            title = wavename+', Stokes V'
            stk = stkv #/stki
            
        i = iline*4 + ipol
        ax[i].plot(wave,stk, marker='+', color='black', linewidth=3, linestyle=' ', alpha=1, label='YES')
        ax[i].set_title(title)
        #ax[i].set_ylim(-0.003,0.003)

fig.tight_layout()
```

## Input: Atmosphere

```python
    atm0,atm1 = readmod(initial_atmos1)
    atm2,atm3 = readmod(initial_atmos2)

    ntau,nparam=atm1.shape

    # 2成分で、CaHと854のStokes Vのred peakを説明したい。
    # Ca 854のStokes-Iの小さな凹みに対応してそう。
    # 2成分目を彩層だけFALCモデルにして磁場を調整した。

    #光球、彩層の切替点
    tau_point = -4.
    pos = (np.where(atm1[:,0] > tau_point))[0]

    # Filling factor
    atm0[1] = 0.9
    atm2[1] = 0.1

    # TEMPERATURE
    atm3[pos,1] = atm1[pos,1]

    # LOS VELOCITY
    #atm1[:,5] = atm1[:,0] + 1e5    # LOS velocity               [cm/s]
    atm3[:,5] = atm3[:,0] + 10e5    # LOS velocity               [cm/s]
    atm3[pos,5] = atm1[pos,5]
    #atm3[margin:ntau-margin,5] = moving_average(atm3[:,5], window_size)

    # MAGNETIC FIELD STRENGTH 
    atm3[:,4] = atm3[:,4]*0.+500.   # Magnetic field strength    [G]
    atm3[pos,4] = atm1[pos,4]*0.+1.

    #atm3[pos,4] = atm1[pos,4]
    #atm3[margin:ntau-margin,4] = moving_average(atm3[:,4], window_size)

    atm3[:,6] = atm3[:,6]*0.+180.          # Magnetic field inclination    [deg]
    atm3[pos,6] = atm1[pos,6]
    #atm3[margin:ntau-margin,6] = moving_average(atm3[:,6], window_size)

    ## Weaken and redirect chromospheric magnetic field
    tmp = atm1[pos,6]
    atm1[:,6] = atm1[:,6]*0.+180.          # Magnetic field inclination    [deg]
    atm1[pos,6] = tmp
    tmp = atm1[pos,4]
    atm1[:,4] = atm1[:,4]*0.5          # Magnetic field inclination    [deg]
    atm1[pos,4] = tmp

    ########################
    # 大気パラメータを０にさせない
    atm3[:,3] = atm1[:,3]*0.+1e3  # Microturbulence            [cm/s]
    atm2[0]   = atm0[0]    # Macroturbulence            [km/s]
    atm3[:,7] = atm1[:,7]*0.+1  # Magnetic field azimuth     [degrees]

    ########################
    # Perturbations


    
    writemod(atm0,atm1,file_atmos)
    writemod(atm2,atm3,file_atmos2)

```
